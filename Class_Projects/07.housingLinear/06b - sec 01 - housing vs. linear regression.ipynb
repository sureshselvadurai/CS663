```python
df['total_bedrooms'] = df.total_bedrooms.fillna(df['total_bedrooms'].median())
```

```python
df.isna().sum()
```

```python
op = pd.get_dummies(df.ocean_proximity,dtype=int)
```

```python
df=  pd.concat([df,op],axis=1)
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Sort correlation values by absolute value, excluding the target column
sorted_corr = corr['median_house_value'].abs().sort_values(ascending=False)[:-1]

print(sorted_corr)
```
